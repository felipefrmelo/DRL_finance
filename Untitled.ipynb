```python
from StockTradingEnv import StockTradingEnv
import pandas as pd
import numpy as np
import random
import torch
import matplotlib.pyplot as plt
import time
import datetime as dt
```

```python
def get_data_alpha(symbols, dates, addSPY=True, colname = ['5. adjusted close']):
    
    from alpha_vantage.timeseries import TimeSeries
    
    ts = TimeSeries(key='ZZXPUWS3V96BN0OV', output_format='pandas')
    
    df = pd.DataFrame(index=dates)                                               
                           
    for symbol in symbols:                           
        df_temp,meta_data = ts.get_daily_adjusted(symbol=symbol, outputsize='full')                       
        df = df.join(df_temp)
        
        if symbol == 'SPY':  # drop dates SPY did not trade                           
            df = df.dropna(subset=["SPY"])                              
    
    return df      
```

```python
start_date = dt.datetime(2012, 1, 1) 
end_date = ed=dt.datetime(2019, 12, 31)
dates = pd.date_range(start_date, end_date)
Symbol = ['PETR4.SA']   
dict_order = {'Symbol': [], 'Order': [], 'Shares': [], 'Index': []}

prices = get_data_alpha(Symbol, dates, False)
prices.dropna(inplace=True)
```

```python
df = prices.rename(columns={column:column.split('.')[1].strip() for column in prices.columns })
```

```python
from ddpg_agent import Agent

from collections import deque

# Create simulation environment
# df = pd.read_csv('./data/AAPL_2018.csv')
env = StockTradingEnv(df)
n_trades = 1000
# Initialize Feed-forward DNNs for Actor and Critic models. 
agent = Agent(state_size=env.observation_space.shape[1] , action_size=3, random_seed=0)

# Set the number of episodes to run the simulation
episodes = 1000
scores = []                        # list containing scores from each episode
scores_window = deque(maxlen=100)  # last 100 scores
best_avg = -np.inf

for episode in range(episodes): 
    # Reset the enviroment
    state = env.reset()
    score = 0
    # set the environment to make transactions
    for i in range(n_trades + 1):
      
        # Predict the best action for the current state. 
        action = agent.act(state, add_noise = True)
        # Action is performed and new state, reward, info are received. 
        new_state, reward, done, info = env.step(np.argmax(action))
        
        # current state, action, reward, new state are stored in the experience replay
        agent.step(state, action, reward, new_state, done)
        
        # roll over new state
        state = new_state
        score += reward      
        if done:
            break
        
    scores_window.append(score)       # save most recent score
    scores.append(score)              # save most recent score
    avg_score = np.mean(scores_window)
    if (episode + 1) % 100 == 0:
        print('\rEpisode {}\tAverage Score: {:.2f}'.format(episode, np.mean(scores_window)))
        env.render()
    if avg_score > best_avg:
        best_avg = avg_score
        torch.save(agent.actor_local.state_dict(), 'checkpoint' + '.pth')    
        if avg_score >= 5080.0:
            print('\nEnvironment solved in {:d} episodes!\tAverage Score: {:.2f}'.format(episode, np.mean(scores_window)))
            torch.save(agent.actor_local.state_dict(), 'checkpoint' + '.pth')

print('\nAverage Implementation Shortfall: ${:,.2f} \n'.format(np.mean(shortfall_hist)))
```

```python
def visualize(info, show_limit=1000):
    history = info['history'][-show_limit:]
    closes = [data[2] for data in history]
    closes_index = [data[1] for data in history]
    # buy tick
    buy_tick = np.array([data[1] for data in history if data[0] == 0])
    buy_price = np.array([data[2] for data in history if data[0] == 0])
    sell_tick = np.array([data[1] for data in history if data[0] == 1])
    sell_price = np.array([data[2] for data in history if data[0] == 1])

    plt.figure(figsize=(14,10)) # change the figure size
    plt.plot(closes_index, closes)
    plt.scatter(buy_tick, buy_price , c='g', marker="^", s=25)
    plt.scatter(sell_tick, sell_price , c='r', marker="v", s=25)
    plt.show(block=True)
    time.sleep(3)
```

```python
visualize(info, show_limit=150)
visualize(info, show_limit=100)
```

```python
# plot cumulative return and max drawdown
def show_mdd(info): # xs is cumulative return / portfolio , if reward u should
    reward_history = [data[4] for data in info['history']]
    df = pd.DataFrame()
    df['reward'] = reward_history
    df['cum_reward'] = df['reward'].cumsum()
    xs = df['cum_reward'].values
    
    i = np.argmax(np.maximum.accumulate(xs) - xs) # end of the period
    j = np.argmax(xs[:i]) # start of period
    plt.figure(figsize=(16,10))
    plt.plot(xs)
    plt.plot([i, j], [xs[i], xs[j]], 'o', color='Red', markersize=10)
    plt.show()
    return xs[-1]
total_return = show_mdd(info)
print("total cumulative return : ",total_return)
print("portfolio value {0} -> {1}" .format(100*10000, info['portfolio']))
```
